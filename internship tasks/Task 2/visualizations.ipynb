# Titanic Data Visualization

## Introduction
In this notebook, we will explore the Titanic dataset using visualizations. We will create bar charts and scatter plots to represent data distributions and relationships.

```python
# importing important libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# loading data set using seaborn
df = pd.read_csv('titanic_cleaned.csv')
```

## Data Overview
Below are the first few rows of the dataset and some basic statistics.

```python
# displaying few rows of the dataset
df.head()
```

```python
# summary of dataset
df.info()
```

```python
# statistical summary of dataset
df.describe()
```

## Visualizations

### Survival Count
The bar chart below shows the count of passengers who survived and did not surviv

```python
plt.figure(figsize=(10, 7))
sns.countplot(x='survived', data=df, palette='viridis')
plt.title('Survived = (1) vs Dead= (0)')
plt.xlabel('Survived')
plt.ylabel('Count')
plt.show()
```

### Passenger Class Distribution
The bar chart below shows the distribution of passengers across different classes.

```python
plt.figure(figsize=(10, 7))
sns.countplot(x='pclass', data=df, palette='viridis')
plt.title("Passenger Class Distribution")
plt.xlabel('Class')
plt.ylabel('Count')
plt.show()
```

### Survival Rate by Gender

```python
plt.figure(figsize=(10, 7))
sns.countplot(data=df, x='sex', hue='survived')
plt.title('Survived vs Dead based on gender')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

### Age Distribution

```python
plt.figure(figsize=(10, 7))
sns.histplot(data=df, x='age', bins=30, kde=True)
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

### Fare Distribution by Class

```python
plt.Figure(figsize=(10, 7))
sns.boxenplot(data=df, x='pclass', y='fare')
plt.title('Fare Distribution based on Class')
plt.xlabel('Class')
plt.ylabel('Fare')
plt.show()
```

### Survival Rate by Class

```python
plt.figure(figsize=(8, 6))
sns.catplot(data=df, x='pclass', hue='survived', kind='count', height=6, aspect=1.5)
plt.title('Survival Count by Class')
plt.xlabel('Class')
plt.ylabel('Count')
plt.show()
```

### Age vs. Fare
The scatter plot below shows the relationship between age and fare paid by passengers.

```python
plt.figure(figsize=(10, 7))
sns.scatterplot(x='age', y='fare', data=df, hue='survived', palette='viridis')
plt.title('Age vs Fare')
plt.xlabel('Age')
plt.ylabel('Fare')
plt.show()
```

### Age vs. Fare by Class
The scatter plot below shows the relationship between age and fare paid by passengers, differentiated by passenger class and survival status.

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='age', y='fare', hue='pclass', style='survived', palette='Set1')
plt.title('Age vs. Fare by Class')
plt.xlabel('Age')
plt.ylabel('Fare')
plt.show()
```

### Pair Plot of Selected Features

```python
selected_features = ['age', 'fare', 'pclass', 'survived']
sns.pairplot(df[selected_features], hue='survived', palette='coolwarm')
plt.suptitle('Pair Plot of Selected Features', y=1.02)
plt.show()
```

### Violin Plot of Age by Survival Status

```python
plt.figure(figsize=(10, 6))
sns.violinplot(data=df, x='survived', y='age', palette='Pastel1')
plt.title('Violin Plot of Age by Survival Status')
plt.xlabel('Survived')
plt.ylabel('Age')
plt.show()
```

### Bar Plot of Embarked Locations

```python
plt.figure(figsize=(8, 6))
sns.countplot(data=df, x='embarked', hue='survived')
plt.title('Survival Count by Embarked Location')
plt.xlabel('Embarked Location')
plt.ylabel('Count')
plt.legend(title='Survived', loc='upper right')
plt.show()
```
